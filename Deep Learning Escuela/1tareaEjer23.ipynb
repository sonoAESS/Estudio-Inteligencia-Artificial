## Operaciones tensoriales y manipulación con NumPy

```python
# importacion de la libreria
import numpy as np
```

### Suma y Broadcasting

```python
#suma elemento a elemento entre matrices del mismo orden
matrix1 = np.random.rand(3, 3)
matrix2 = np.random.rand(3, 3)
print("Suma elemento a elemento:\n", matrix1 + matrix2)
```

```python
# Suma mediante broadcasting (matriz + vector), expande dimensiones de la estructura menor 
vector = np.array([1, 2, 3])
print("Suma con broadcasting:\n", matrix1 + vector)
```

### Multiplicación y producto de Hadamard

```python
#Multiplicación matricial, cantidad de columnas de primera coincide con filas de la segunda
print("Multiplicación matricial:\n", np.dot(matrix1, matrix2))
```

```python
#Producto de Hadamard, multiplicacion de posicion por posicion
print("Producto de Hadamard:\n", matrix1 * matrix2)
```

### Transposición de Tensores

```python
#Transposición de tensores 2D 
tensor2d = np.random.rand(4, 5)
print("Transposición 2D:\n", tensor2d.T)
#como una matriz la transpuesta de m*n es n*m
#La transposición de tensores 2D se logra cambiando el orden de los índices, es decir, el índice de fila se convierte en el índice de columna. Un truco es sustituir el elmento en la posición 1, 2 por el de la posición 2, 1.
```

```python
#transposición de tensores de 3 dimensiones
tensor3d = np.random.rand(3, 4, 5)
print("Transposición 3D:\n", np.transpose(tensor3d, axes=(1, 0, 2)))
#un tensor de tamaño l*m*n se suele emplear la transposición parcial, se intercambian solo dos de las tres dimensiones, como por ejemplo m*l*n
#en este caso sustituimos el primer eje con el segundo
```

## Implementación y análisis de funciones de activación

```python
#libreria
import matplotlib.pyplot as plt
#constante para los ejemplos
x_values = np.linspace(-10, 10, 100)
```

### Función Sigmoide

Conocida como función logística, su fórmula es la división de 1 con 1 + euler elevado a la -x. Siendo x el valor introducido.

Se suele emplear en problemas de clasificación binaria. Su salida se puede interpretar como una probabilidad.

El rango de salida es entre 0 y 1.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

plt.plot(x_values, sigmoid(x_values), label="Sigmoide")
plt.title("Función Sigmoide")
plt.xlabel("x")
plt.ylabel("Sigmoide(x)")
plt.legend()
plt.show()
```

### Función ReLu

Su fórmula es el máximo valor entre 0 y x. Es decir, con un rango de salida de 0 a infinito se tiene que solo son adecuadas las entradas positivas.

Se emplea en redes profundas, para evitar la saturación.

```python
def relu(x):
    return np.maximum(0, x)

plt.plot(x_values, relu(x_values), label="ReLU")
plt.title("Función ReLU")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.legend()
plt.show()
```

### Función Tangente Hiperbólica (Tanh)

Esta función es similar a la sigmoide pero con un rango de salida distinto, desde -1 a 1. Su fórmula es 2 dividido por 1 + eular elevado a -2x, todo eso - 1. Se emplea principalmente en RNN.

```python
def tanh(x):
    return np.tanh(x)

plt.plot(x_values, tanh(x_values), label="Tanh")
plt.title("Función Tanh")
plt.xlabel("x")
plt.ylabel("Tanh(x)")
plt.legend()
plt.show()
```
