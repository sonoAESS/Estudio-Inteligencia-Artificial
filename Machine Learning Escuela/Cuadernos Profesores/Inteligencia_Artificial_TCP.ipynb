Existen métodos para la normalización de variables como los siguientes:
- Estandarización: donde se busca que las variables tengan una media 0 y una desviación estándar de 1. Se resta a cada valor la media y se divide entre la desviación estándar.
- Escalado Min-Max: ajusta los valores a un rango específico, generalmente de 0 a 1. Es sensible a valores atípicos por lo que no se emplea mucho. Conciste en restar al valor original su mínimo y dividir esto entre la resta del máximo con el mínimo.
- Transformación logarítmica: reduce el rango e impacto de los outliers, se emplea principalmente en distribuciones sesgadas o exponenciales. Consiste en obtener el logarítmo natural en base 10 de la variable.

### 2.6 Se conoce que dos variables estan fuertemente relacionadas cuando poseen un índice de correlación de pearlson mayor a 75%. Hay variables que se relacionen fuertemente?, de haberlas, indique cuales.

```python
df1['diagnosis'].value_counts() # B y M
df1['diagnosis']=df1['diagnosis'].map({'M':0,'B':1})#hacerla numerica para matriz de correlacion
```

```python
from sklearn.preprocessing import StandardScaler
#estandarizar antes de obtener matriz de correlacion
scaler=StandardScaler()
df1_scaled=scaler.fit_transform(df1)
df1_scaled=pd.DataFrame(df1_scaled,columns=df1.columns)

#eliminar columna id por ser innecesaria
df1S=df1_scaled.drop(columns=['id'])
df1S.head()
```

```python
corrMat=df1S.corr(method='pearson')
plt.figure(figsize=(16, 12))
sns.heatmap(corrMat,annot=True,cmap='coolwarm')
plt.show()
```

```python
#mapa de calor con correlaciones de mas de 0.75
strongC=corrMat[(corrMat.abs()>0.75)&(corrMat!=1)]
sns.heatmap(strongC)
plt.show()
```
