## Linear Regression
En este ejemplo generamos datos con una relación lineal entre velocidad de la página y el amount purchased

```python
import numpy as np
import matplotlib.pyplot as plt

pageSpeeds=np.random.normal(3.0,2.0,1000)
purchaseAmount=100-(pageSpeeds+np.random.normal(0,0.1,1000))*3 #distribucion normal

plt.scatter(pageSpeeds, purchaseAmount)
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err=stats.linregress(pageSpeeds,purchaseAmount)
```

```python
r_value**2
```

```python
def predict(x):
    return slope*x+intercept

fitline=predict(pageSpeeds)

plt.scatter(pageSpeeds,purchaseAmount)
plt.plot(pageSpeeds,fitline,c='r')
plt.show()
```
